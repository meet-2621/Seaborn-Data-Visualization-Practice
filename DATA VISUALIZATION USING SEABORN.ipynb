#SCATTER PLOT

```python
sns.scatterplot(data = penguins , x = 'flipper_length_mm' , y = 'body_mass_g' , hue = 'island')
```

```python
sns.scatterplot(data = penguins , x = 'flipper_length_mm' , y = 'body_mass_g' , hue = 'sex')
```

```python
sns.set_style('dark')
sns.scatterplot(data = penguins , x = 'flipper_length_mm' , y = 'body_mass_g' , hue = 'species')
sns.despine(left = True)
```

```python
sns.set_style('ticks')
sns.scatterplot(data = penguins , x = 'flipper_length_mm' , y = 'body_mass_g' , hue = 'species')
sns.despine(left = True)    # y axis got removed!!
```

```python
sns.set_style('ticks')
#sns.set_context('poster')  
#sns.set_context('talk') 
#sns.set_context('notebook') 
sns.set_context('paper') 
sns.scatterplot(data = penguins , x = 'flipper_length_mm' , y = 'body_mass_g' , hue = 'species' , palette = "Dark2")  # "Set1" , "Set2" , "Set3"
sns.despine(left = True) 
```

```python
sns.scatterplot(data = penguins , x = 'species' , y = 'body_mass_g' , hue = 'island' , style = 'sex' , alpha = 0.5)
```

#STRIP PLOT

```python
sns.stripplot(data = penguins , x = 'species' , y = 'body_mass_g' , hue = 'island')
```

```python
sns.stripplot(data = penguins , x = 'species' , y = 'body_mass_g' , hue = 'island' , dodge = True , jitter = True)
```

#SWARMPLOT

```python
sns.set_context('talk')
sns.swarmplot(data = penguins , x = 'species' , y = 'body_mass_g' , hue = 'island')
sns.despine()
```

#HISTOGRAM

```python
sns.set_context('notebook')
sns.histplot(data = penguins , x = 'body_mass_g' , hue = 'sex' )#, multiple = 'stack')
sns.despine()
```

#REGRESSION PLOT

```python
sns.set_context('notebook')
sns.regplot(data = penguins , x = 'body_mass_g' ,y = 'flipper_length_mm' , color = 'gray' , ) # , scatter = False
sns.despine()
```

#LINE PLOT

```python
sns.set_style('whitegrid')
sns.lineplot(data = penguins , x = 'body_mass_g' ,y = 'flipper_length_mm' , hue = 'sex' , style = 'island') 
sns.despine()
```

#JOINTPLOTS

```python
sns.set_style('whitegrid')
sns.jointplot(data = penguins , x = 'body_mass_g' ,y = 'flipper_length_mm' , color = 'brown' , kind = 'reg') #, kind = 'kde' , 'scatter' , 'reg' 
sns.despine()
```

#BARPLOT

```python
sns.set_style('whitegrid')
sns.barplot(data = penguins , x = 'species' ,y = 'body_mass_g' , hue = 'sex' , palette = ['pink' , 'skyblue'] , estimator = np.var)
#np.var , np.std , np.median , np.sum
sns.despine() 
```

#COUNT PLOT

```python
sns.set_style('whitegrid')
sns.countplot(data = penguins , x = 'species' , hue = 'island' , palette = "Set2")
sns.despine() 
```

#BOXPLOT

```python
sns.set_style('whitegrid')
sns.boxplot(data = penguins , x = 'island' ,y = 'body_mass_g' , hue = 'species' , palette = "Set3")
sns.despine() 
```

violinplot

```python
sns.set_style('whitegrid')
sns.violinplot(data = penguins , x = 'species' ,y = 'body_mass_g' , hue = 'island' , palette = "Set3" , split = 'True')
sns.despine() 

#inner box ---->>> Inter Quartile Range
```

```python
sns.set_style('whitegrid')
sns.violinplot(data = penguins , x = 'species' ,y = 'body_mass_g' )
sns.swarmplot(data = penguins , x = 'species' ,y = 'body_mass_g' , color = 'black' , size = 3)
sns.despine()
```

```python
columns = ['bill_length_mm'	, 'bill_depth_mm' ,	'flipper_length_mm' , 'body_mass_g']
sns.heatmap(data = penguins[columns].corr() )
```

```python
columns = ['bill_length_mm'	, 'bill_depth_mm' ,	'flipper_length_mm' , 'body_mass_g']
sns.heatmap(data = penguins[columns].corr() , annot = True , cmap = "coolwarm")
```
